```python
# import libraries we will need as it's alias 

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read the CSV as a DataFrame
office_episodes=pd.read_csv("datasets/office_episodes.csv")

# print the first 10 row of the Dataframe
office_episodes.head(10)
```

```python
# create a sizing list 
marker_size=[]
for lab,row in office_episodes.iterrows():
  if row["has_guests"]==False:
    marker_size.append(25)
  else:
    marker_size.append(250)
    
# print the first 10 values in the list 
marker_size[:10]
```

```python
# create a color_scheme list 
colors=[]
for lab,row in office_episodes.iterrows():
  if row["scaled_ratings"]<.25:
    colors.append("red")
  elif row["scaled_ratings"]>=.25 and row["scaled_ratings"]<.50:
     colors.append("orange")
  elif row["scaled_ratings"]>=.50 and row["scaled_ratings"]<.75:
    colors.append("lightgreen")
  else:
    colors.append("darkgreen")
    
# print the first 10 values in the list 
colors[:10]
```

```python
# Create a fig and set it's size 
fig = plt.figure(figsize=(11, 7))

# Create a scatter plot
plt.scatter(office_episodes["episode_number"], office_episodes["viewership_mil"], c = colors, s = marker_size,marker="*")

# Create a Title 
plt.title("Popularity, Quality, and Guest Appearances on the Office",size=20)

# Create a x-axis & y-axis label 
plt.xlabel("Episode Number",size=16)
plt.ylabel("Viewership (Millions)",size=16)
# show the plot 
plt.show()
```
